# CSVtoJSON File Converter

```python
# A fuction to convert CSV file to JSON file
def csv_toJson(csv_path,json_path): # csv file path and location where JSON file has to be stored. 
    
    #Import essential libraries
    import csv
    import json
    import pandas as pd
    
    # Reading data
    data=pd.read_csv(csv_path)
    
    # Trasforming the data into required format
    t=data.T
    t.index = t.index.set_names('company')
    t = t.reset_index()
    t.to_csv("output.csv")
    t_data = pd.read_csv('output.csv')
    
    #splitting the data
    #Split 1
    t1 = t_data.iloc[0:4,:]

    t1.to_csv("t1.csv")
    #Split 2 "policyInformation":
    t2 =t_data.iloc[4:6,:]
    t2=t2.T.iloc[2:4,:]
    t2.to_csv("t2.csv")
    #Split 3 "insuredPersonInformation":
    #      "insuredPersonInformationList":
    #          "person":
    tt= t_data.iloc[6:10,:]
    tt=tt.T.iloc[2:4,:]
    tt.to_csv("tt.csv")
    #Split 4          "primaryAddress":
    t4 =t_data.iloc[10:14,:]
    t4=t4.T.iloc[2:4,:]
    t4.to_csv("t4.csv")
    #Split 5 "insuredCoverages":
    #      "insuredCoverage":
    t5 =t_data.iloc[14:16,:]
    t5=t5.T.iloc[2:4,:]
    t5.to_csv("t5.csv")
    #Split 6           "planCode":
    t6 =t_data.iloc[16:17,:]
    t6=t6.T.iloc[2:4,:]
    t6.to_csv("t6.csv")
    #Split 7 
    t7 =t_data.iloc[17:20,:]
    t7=t7.T.iloc[2:4,:]
    t7.to_csv("t7.csv")
    #Split 8           "sumTotalCoverageAmount":
    t8 =t_data.iloc[20:21,:]
    t8=t8.T.iloc[2:4,:]
    t8.to_csv("t8.csv")
    #Split 9 
    t9 =t_data.iloc[22:23,:]
    t9=t9.T.iloc[2:4,:]
    t9.to_csv("t9.csv")

    #Split 10
    t10 =t_data.iloc[23:,:]
    t10=t10.T.iloc[2:4,:]
    t10.to_csv("t10.csv")
    
    # create a dictionary 
    data = {}
    
    # Open a csv reader called Reader for each splitted data
    t1=open('t1.csv','r')
    reader1 = csv.reader(t1)
    for row in reader1:
        data[row[2]] = {row[3],row[4]}

    t2=open('t2.csv','r')
    reader2 = csv.reader(t2)
    for row in reader2:
        data["policyInformation"] = {"Number":row[1],"companyCode":row[2],"insuredPersonInformation":{"insuredPersonInformationList":{"person"}}}


    tt=open('tt.csv','r')
    readertt = csv.reader(tt)
    for row in readertt:
        data["policyInformation"]["insuredPersonInformation"]["insuredPersonInformationList"] ={"person":{"personNumber":row[1],"nameComponents":{"nameType":row[2],'firstName':row[3],"lastName":row[4]}}}

    t5=open('t5.csv','r')
    reader5 = csv.reader(t5)
    data["policyInformation"]["insuredPersonInformation"]["insuredPersonInformationList"]["insuredCoverages"] = {"insuredCoverage"}
    for row in reader5:
        data["policyInformation"]["insuredPersonInformation"]["insuredPersonInformationList"]["insuredCoverages"]={"insuredCoverage":{"coverageRecordNumber":row[1],"insuredRelation":row[2]}}
    data["policyInformation"]["insuredPersonInformation"]["insuredPersonInformationList"]["insuredCoverages"]["insuredCoverage"]={"planCode"}

    t6=open('t6.csv','r')
    reader6 = csv.reader(t6)
    for row in reader6:
        data["policyInformation"]["insuredPersonInformation"]["insuredPersonInformationList"]["insuredCoverages"]["insuredCoverage"]={"planCode":{'code':row[1]}}


    t7=open('t7.csv','r')
    reader7 = csv.reader(t7)
    for row in reader7:
        data["policyInformation"]["insuredPersonInformation"]["insuredPersonInformationList"]["insuredCoverages"]["insuredCoverage"]={'productCode':row[1],'productDescription':row[2],'baseRiderIndicator':row[3]}


    t8=open('t8.csv','r')
    reader8 = csv.reader(t8)
    for row in reader8:
        data["policyInformation"]["insuredPersonInformation"]["insuredPersonInformationList"]["insuredCoverages"]["insuredCoverage"]["inflatedBenefitAmount"]={"inflatedBenefitAmount":row[1]}

    t9=open('t9.csv','r')
    reader9 = csv.reader(t9)
    for row in reader9:
        data["policyInformation"]["insuredPersonInformation"]["insuredPersonInformationList"]["insuredCoverages"]["insuredCoverage"]["premiumClass"]={"code":row[1]}

    t10=open('t10.csv','r')
    reader10 = csv.reader(t10)
    for row in reader10:
        data["policyInformation"]["insuredPersonInformation"]["insuredPersonInformationList"]["insuredCoverages"]["insuredCoverage"]["premiumClass"]={"coverageStatus":row[1],"singlePayAmount":row[2],"preArrangedBenefit":row[3],"guaranteedBenefitAmount":row[4],"increaseType": row[5]}
    
    def set_default(obj):
        if isinstance(obj, set):
            return list(obj)
        raise TypeError
    # Closing all the opened files
    t1.close()
    t2.close()
    tt.close()
    t5.close()
    t6.close()
    t7.close()
    t8.close()
    t9.close()
    t10.close()
    # function to dump data 
    with open(json_path, 'w', encoding='utf-8') as jsonf: 
        jsonf.write(json.dumps(data, default=set_default,indent=4)) # Saving output as JSON file
```
