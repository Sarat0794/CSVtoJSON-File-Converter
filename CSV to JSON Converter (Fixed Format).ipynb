```python
def csvToJSON(csv_path,json_path):    
    # Import essential libraries
    import pandas as pd
    from itertools import groupby 
    from collections import OrderedDict
    import json 

    # Read data
    # "dtype=str":everything has to be str type which can be serialized to JSON.
    df=pd.read_csv(csv_path, dtype=str) 
    df=df.fillna("")
    # create an emply list
    output=[]

    # Track the number of iterations
    count=0 

    for (applicationSignatureDate,applicationReceivedDate,coverageChangeEffectiveDate,getCoverageMaintenanceToken), bag in df.groupby(["applicationSignatureDate","applicationReceivedDate","coverageChangeEffectiveDate","getCoverageMaintenanceToken"]):

        output2=[]

        contents_df = bag.drop(["applicationSignatureDate","applicationReceivedDate","coverageChangeEffectiveDate","getCoverageMaintenanceToken"],axis=1)

        for (Number,companyCode), bag in df.iloc[count:,4:].groupby(["Number","companyCode"]):

            contents_2 = bag.drop(["Number","companyCode"],axis=1)

            insuredPersonInformation={"insuredPersonInformationList":[{"person":[OrderedDict(row) for i,row in contents_2.iterrows()],"insuredCoverages":{"insuredCoverage":[OrderedDict(row) for i,row in contents_2.iterrows()]}}]}

            output2.append(OrderedDict([("Number",Number),("companyCode",companyCode),("insuredPersonInformation",insuredPersonInformation)]))

            policyInformation = output2

        count+=1

        output.append(OrderedDict([("applicationSignatureDate",applicationSignatureDate),("applicationReceivedDate",applicationReceivedDate),("coverageChangeEffectiveDate",coverageChangeEffectiveDate),("getCoverageMaintenanceToken",getCoverageMaintenanceToken),("policyInformation",policyInformation)]))
    for i in range(len(df)):
        output[i]['policyInformation']=output[i]['policyInformation'][0]
    for i in range(len(output)):

        output[i]['policyInformation']["insuredPersonInformation"]["insuredPersonInformationList"][0]["person"][0]["nameComponents"]={key: output[i]['policyInformation']["insuredPersonInformation"]["insuredPersonInformationList"][0]["person"][0][key] for key in output[0]['policyInformation']["insuredPersonInformation"]["insuredPersonInformationList"][0]["person"][0].keys() 
                                       & {'nameType', 'firstName','lastName'}}
    for i in range(len(output)):

        output[i]['policyInformation']["insuredPersonInformation"]["insuredPersonInformationList"][0]["person"][0]["primaryAddress"]={key: output[i]['policyInformation']["insuredPersonInformation"]["insuredPersonInformationList"][0]["person"][0][key] for key in output[0]['policyInformation']["insuredPersonInformation"]["insuredPersonInformationList"][0]["person"][0].keys() 
                                       & {'addressLine1', 'cityName','stateCode','zipCode'}}




    for i in range(len(output)):

        output[i]['policyInformation']["insuredPersonInformation"]["insuredPersonInformationList"][0]["insuredCoverages"]["insuredCoverage"][0]["planCode"]={key: output[i]['policyInformation']["insuredPersonInformation"]["insuredPersonInformationList"][0]["insuredCoverages"]["insuredCoverage"][0][key] for key in output[0]['policyInformation']["insuredPersonInformation"]["insuredPersonInformationList"][0]["insuredCoverages"]["insuredCoverage"][0].keys() 
                                       & {"code"}}

        output[i]['policyInformation']["insuredPersonInformation"]["insuredPersonInformationList"][0]["insuredCoverages"]["insuredCoverage"][0]["sumTotalCoverageAmount"]={key: output[i]['policyInformation']["insuredPersonInformation"]["insuredPersonInformationList"][0]["insuredCoverages"]["insuredCoverage"][0][key] for key in output[0]['policyInformation']["insuredPersonInformation"]["insuredPersonInformationList"][0]["insuredCoverages"]["insuredCoverage"][0].keys() 
                                       & {"amt"}}


    p_duplicates=["nameType","firstName","lastName","code","addressLine1","cityName","zipCode","coverageRecordNumber","insuredRelation","amt","productCode","productDescription","baseRiderIndicator","inflatedBenefitAmount","premiumClass","coverageStatus","singlePayAmount","preArrangedBenefit","guaranteedBenefitAmount","increaseType","stateCode"]
    i_duplicates=["nameType","firstName","lastName","addressLine1","cityName","zipCode","stateCode","personNumber","code","amt"]
    for i in range(len(output)):

        for key in p_duplicates:

            #del output[i]["policyInformation"][0]["insuredPersonInformation"]["insuredPersonInformationList"][0]["person"][0][key]
            del output[i]['policyInformation']['insuredPersonInformation']["insuredPersonInformationList"][0]['person'][0][key]

        for key in i_duplicates:
            del output[i]['policyInformation']['insuredPersonInformation']["insuredPersonInformationList"][0]["insuredCoverages"]["insuredCoverage"][0][key]
    # Saving output to JSON file with filename ExpectedJsonFile.json
    print (json.dumps(output, indent=4))
    with open(json_path, 'w') as outfile:
        outfile.write(json.dumps(output, indent=4))

```
