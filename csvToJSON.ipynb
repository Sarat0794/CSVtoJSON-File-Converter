# CSV to JSON Converter

```python
# A fuction to convert CSV file to JSON file
# Parameters:csv file path location and JSON file location, respectively. 
def csvToJSON(csv_path,json_path): 
    
    # Import essential libraries
    import pandas as pd
    from itertools import groupby 
    from collections import OrderedDict
    import json 

    # Read data
    # "dtype=str":everything has to be str type which can be serialized to JSON.
    df=pd.read_csv(csv_path, dtype=str) 
    
    # create an emply list
    output=[]
    
    # Track the number of iterations
    count=0 
    
    for (applicationSignatureDate,applicationReceivedDate,coverageChangeEffectiveDate,getCoverageMaintenanceToken), bag in df.groupby(["applicationSignatureDate","applicationReceivedDate","coverageChangeEffectiveDate","getCoverageMaintenanceToken"]):
        
        output2=[]
        
        contents_df = bag.drop(["applicationSignatureDate","applicationReceivedDate","coverageChangeEffectiveDate","getCoverageMaintenanceToken"],axis=1)
        
        for (Number,companyCode), bag in df.iloc[count:,4:].groupby(["Number","companyCode"]):
            
            contents_2 = bag.drop(["Number","companyCode"],axis=1)
            
            insuredPersonInformation={"insuredPersonInformationList":[{"person":[OrderedDict(row) for i,row in contents_2.iterrows()]}]}
            
            output2.append(OrderedDict([("Number",Number),("companyCode",companyCode),("insuredPersonInformation",insuredPersonInformation)]))

            policyInformation = [output2]

        count+=1
        
        output.append(OrderedDict([("applicationSignatureDate",applicationSignatureDate),("applicationReceivedDate",applicationReceivedDate),("coverageChangeEffectiveDate",coverageChangeEffectiveDate),("getCoverageMaintenanceToken",getCoverageMaintenanceToken),("policyInformation",policyInformation)]))
        
        for i in range(len(output)):
            
            output[i]['policyInformation'][0][0]["insuredPersonInformation"]["insuredPersonInformationList"][0]["person"][0]["nameComponents"]={key: output[0]['policyInformation'][0][0]["insuredPersonInformation"]["insuredPersonInformationList"][0]["person"][0][key] for key in output[0]['policyInformation'][0][0]["insuredPersonInformation"]["insuredPersonInformationList"][0]["person"][0].keys() 
                                   & {'nameType', 'firstName','lastName'}}
        for i in range(len(output)):
           
            output[i]['policyInformation'][0][0]["insuredPersonInformation"]["insuredPersonInformationList"][0]["person"][0]["primaryAddress"]={key: output[0]['policyInformation'][0][0]["insuredPersonInformation"]["insuredPersonInformationList"][0]["person"][0][key] for key in output[0]['policyInformation'][0][0]["insuredPersonInformation"]["insuredPersonInformationList"][0]["person"][0].keys() 
                                   & {'addressLine1', 'cityName','stateCode','zipCode'}}

        #Splitting data based on the given JSON format.
        t1= df.iloc[:,14:16]
        t2= df.iloc[:,16:17]
        t3= df.iloc[:,17:20]
        t4= df.iloc[:,20:21]
        t5= df.iloc[:,21:]

        for i in range(len(output)):
            output[i]["policyInformation"][0][0]["insuredPersonInformation"]["insuredPersonInformationList"][0]["insuredCoverages"]={"insuredCoverage":[{k:list(t1[k].unique()) for k in t1.columns }]}
            output[i]["policyInformation"][0][0]["insuredPersonInformation"]["insuredPersonInformationList"][0]["insuredCoverages"]["insuredCoverage"][0]["planCode"]={k:list(t2[k].unique()) for k in t2.columns }
            output[i]["policyInformation"][0][0]["insuredPersonInformation"]["insuredPersonInformationList"][0]["insuredCoverages"]["insuredCoverage"].append({k:list(t3[k].unique()) for k in t3.columns})
            output[i]["policyInformation"][0][0]["insuredPersonInformation"]["insuredPersonInformationList"][0]["insuredCoverages"]["insuredCoverage"][0]["sumTotalCoverageAmount"]={k:list(t4[k].unique()) for k in t4.columns }
            output[i]["policyInformation"][0][0]["insuredPersonInformation"]["insuredPersonInformationList"][0]["insuredCoverages"]["insuredCoverage"].append({k:list(t5[k].unique()) for k in t5.columns})
    
    # Removing duplicates
    for i in range(1,len(output[0]['policyInformation'][0])):
        
        del output[0]['policyInformation'][0][i]
    
    duplicates=["nameType","firstName","lastName","code","addressLine1","cityName","zipCode","coverageRecordNumber","insuredRelation","amt","productCode","productDescription","baseRiderIndicator","inflatedBenefitAmount","premiumClass","coverageStatus","singlePayAmount","preArrangedBenefit","guaranteedBenefitAmount","increaseType","stateCode"]
    
    for i in range(len(output)):
        
        for key in duplicates:
            
            #del output[i]["policyInformation"][0]["insuredPersonInformation"]["insuredPersonInformationList"][0]["person"][0][key]
            del output[i]['policyInformation'][0][0]['insuredPersonInformation']["insuredPersonInformationList"][0]['person'][0][key]
    
    # Saving output to JSON file with filename ExpectedJsonFile.json
    print (json.dumps(output, indent=4))
    with open('ExpectedJsonFile.json', 'w') as outfile:
        outfile.write(json.dumps(output, indent=4))
```
